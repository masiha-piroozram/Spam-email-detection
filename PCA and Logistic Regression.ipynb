```python
# Import packages
import pandas as pd # to load the dataframe
from sklearn.preprocessing import StandardScaler # to standardize the features
from sklearn.decomposition import PCA # to apply PCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
#import the dataset
data = pd.read_csv('spambase.csv', header=None)
data_copy = data
data_copy = data_copy.drop(57, axis='columns')
data_copy
```

```python
#Standardize the features
#Create an object of StandardScaler which is present in sklearn.preprocessing
scalar = StandardScaler()
std_data = pd.DataFrame(scalar.fit_transform(data_copy)) #scaling the data
std_data
```

```python
#Applying PCA
#Taking no. of Principal Components as 10
pca = PCA(n_components = 10)
pca.fit(std_data)
data_pca = pca.transform(std_data)
data_pca = pd.DataFrame(data_pca)
data_pca[10] = data.iloc[:, 57]
data_pca
```

```python
#split the dataset into train set and test set
x_train, x_test, y_train, y_test = train_test_split(data_pca.iloc[:, 0:10],data_pca.iloc[:, 10],test_size = .25)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='liblinear', penalty='l1')
model.fit(x_train, y_train)
predict = model.predict(x_test)
accuracy_score(y_test,predict)
```

```python
confusion_matrix(y_test,predict)
```
